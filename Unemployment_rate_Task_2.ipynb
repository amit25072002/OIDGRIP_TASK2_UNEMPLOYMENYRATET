```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
data = pd.read_csv('/content/drive/MyDrive/task 2 database/unemployment dataset.zip')
print(data.head())
```

```python
data.columns = ["Region","date","frequency",
                "Estimated unemployment",
                "Estimated Employed",
                "labour rate",
                "area"]
 #plt.style.use('sns.whitebrid')
plt.figure(figsize=(8,6))
sns.heatmap(data.corr())
plt.show()
```
